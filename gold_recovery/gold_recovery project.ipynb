В нашем технологическом процессе есть три этапа, плюс финальный выход. Для каждого этапа доступны такие типы параметров, как вход, выход, состояние и расчеты. Я сравнила, что доступно в обучающем датасете, а что в тестовом.

Информация о входе (флотация и первая очистка) и состоянии установок (все три этапа) одинаково доступна в обоих датасетах. 

Информация о выходе доступна только в обучающем датасете. Это содержание элементов в концентрате и хвостах на трех этапах, плюс итоговые хвосты - я предполагаю, что это некоторый итог по хвостам со всех этапов. Нужно будет проверить, как связаны значения

Для первого этапа есть расчеты, но только в обучающем датасете. Из названий кажется, что это отношения:
- сульфата к золоту в концентрате
- сульфата отдельно по каждой установке к золоту на входе
- золота к свинцу (непонятно, на входе или на выходе)

Я думаю, что эти расчеты здесь не случайно. Вероятно, эти параметры пригодятся модели для прогноза. Может быть, если понять, как они считаются, их можно восстановить для тестовой выборки

### Проверка эффективности обогащения на флотации

```python
# добавим столбец и запишем в него эффективность обогащения, рассчитанную по формуле
# (C*(F-T)) / (F*(C-T)) * 100
# C - rougher.output.concentrate_au
# F - rougher.input.feed_au
# T - rougher.output.tail_au

def recovery(row):
    conc = row['rougher.output.concentrate_au']
    feed = row['rougher.input.feed_au']
    tail = row['rougher.output.tail_au']
    return((conc*(feed-tail))/(feed*(conc-tail))*100)

train_data['rougher.calculation.recovery'] = train_data.apply(recovery, axis=1)

# посмотрим на recovery из файла и расчетное
train_data[['rougher.output.recovery', 'rougher.calculation.recovery']].head(10)
```

```python
# значения совпадают. Для уверенности рассчитаем MAE
mean_absolute_error(train_data['rougher.output.recovery'], train_data['rougher.calculation.recovery'])
```

Ошибка минимальна. Эффективность обогащения по формуле рассчитана верно

```python
# аналогично для final recovery, только используем векторные вычисления
c = 'final.output.concentrate_au'
f = 'rougher.input.feed_au'
t = 'final.output.tail_au'

train_data['final.calculation.recovery'] = (train_data[c]*(train_data[f]-train_data[t])) / (train_data[f]*(train_data[c]-train_data[t])) * 100
train_data[['final.output.recovery', 'final.calculation.recovery']].head(10)
```

### sMAPE, метрика качества

```python
def smape(y_trues, y_preds):
    return(100*((abs(y_trues-y_preds))/((abs(y_trues)+abs(y_preds))/2)).mean())

a = pd.Series([2, 4, 9, 8, 5, 5])
b = pd.Series([3, 4, 8, 5, 4, 5])

print(smape(a, b))    # проверила на калькуляторе, ок
```

```python
def final_smape(y_trues_rougher, y_preds_rougher, y_trues_final, y_preds_final):
    return(0.25*smape(y_trues_rougher, y_preds_rougher) + 0.75*smape(y_trues_final, y_preds_final))
```
